```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ima = pd.read_excel('/Users/cyano/Desktop/tsi_fui.xlsx').to_numpy()
```

```python
x = ima[:,0]
y = ima[:,1]
```

```python
# 每个点的颜色
colors = []
for i in x:
    if i <= 30:
        colors.append('b')
    elif 30 < i <= 50:
        colors.append('y')
    else:
        colors.append('r')
```

```python
fig,ax = plt.subplots()
ax.scatter(x,y,c = colors,alpha = 0.4,s =15)
plt.plot(np.ones((475,))*30,y,'--')
plt.plot(np.ones((475,))*50,y,'--')
plt.xlabel('TSI')
plt.ylabel('FUI')
plt.title('The ralationship of TSI and FUI')
plt.axis([0,100,0,22])
x1 = np.linspace(0,66,700)
y1 = 2.1912 * np.exp(0.0314 * x1)
plt.plot(x1,y1,'C5')
```

```python
from scipy.optimize import curve_fit
def func(x,a,b):
    return a * np.exp(b * x)
```

```python
popt,pcov = curve_fit(func,x,y)
popt
```
