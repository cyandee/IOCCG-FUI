```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gt = pd.read_excel('/Users/cyano/Desktop/gt_fui.xlsx')
```

```python
gtr = gt.to_numpy()
```

```python
tsi = gtr[:,0]
fui = gtr[:,1]
# 每个点的颜色
colors = []
for i in tsi:
    if i <= 30:
        colors.append('b')
    else:
        colors.append('y')
```

```python
fig,ax = plt.subplots()
ax.scatter(tsi,fui,alpha = 0.6,c = colors,s = 10)
plt.xlabel('TSI derived from in situ Chla concentrations')
plt.ylabel('FUI from concurrent Sentinel-2')
plt.axis([10,55,5,15])
y2  = np.arange(0,22)
x2 = np.ones(22,)*30
x3 = np.ones(22,)*50
x4 = np.linspace(18,42,500)
y4 = 4.3058*np.exp(0.0268*x4)
plt.plot(x2,y2,'C2--',x3,y2,'C8--',x4,y4,'C3',lw = 2)
plt.text(12,13.5,r'$y = 4.3058*e^{0.0268x}$')
plt.text(12,12.5,r'$R^2 = 0.693$')
plt.savefig('/Users/cyano/Desktop/gtfu.jpg',dpi = 300)
```
